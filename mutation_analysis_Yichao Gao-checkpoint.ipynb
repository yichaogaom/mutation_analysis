```python
import os
```

```python
import pandas as pd
```

```python
folder = r"C:\Users\24214\Downloads\mutation_notebook_example\data"
data_list = []
```

```python
for filename in os.listdir(folder):
    if filename.endswith(".txt"):
        file_path = os.path.join(folder, filename)
        df = pd.read_csv(file_path, sep="\t")
        data_list.append(df)
```

```python
df_all = pd.concat(data_list, ignore_index=True)
print("Shape of combined data:", df_all.shape)
df_all.head()
```

```python
df_all['mRNA_WT_mean'] = df_all[['mRNA.Expression.WT.Rep1',
                                 'mRNA.Expression.WT.Rep2',
                                 'mRNA.Expression.WT.Rep3']].mean(axis=1)

df_all['mRNA_Mut_mean'] = df_all[['mRNA.Expression.Mut.Rep1',
                                  'mRNA.Expression.Mut.Rep2',
                                  'mRNA.Expression.Mut.Rep3']].mean(axis=1)
```

```python
df_all['mRNA_change'] = df_all['mRNA_Mut_mean'] - df_all['mRNA_WT_mean']
```

```python
df_all['Protein_WT_mean'] = df_all[['Protein.Expression.WT.Rep1',
                                    'Protein.Expression.WT.Rep2',
                                    'Protein.Expression.WT.Rep3']].mean(axis=1)

df_all['Protein_Mut_mean'] = df_all[['Protein.Expression.Mut.Rep1',
                                     'Protein.Expression.Mut.Rep2',
                                     'Protein.Expression.Mut.Rep3']].mean(axis=1)
```

```python
df_all['Protein_change'] = df_all['Protein_Mut_mean'] - df_all['Protein_WT_mean']
```

```python
df_all['Viability_WT_mean'] = df_all[['CellViability.WT.Rep1',
                                      'CellViability.WT.Rep2',
                                      'CellViability.WT.Rep3']].mean(axis=1)

df_all['Viability_Mut_mean'] = df_all[['CellViability.Mut.Rep1',
                                       'CellViability.Mut.Rep2',
                                       'CellViability.Mut.Rep3']].mean(axis=1)
```

```python
df_all['Viability_change'] = df_all['Viability_Mut_mean'] - df_all['Viability_WT_mean']
```

```python
df_all[['Gene', 'mRNA_change', 'Protein_change', 'Viability_change']].head()
```

```python
mutation_types = []
mutation_regions = []
```

```python
for i in range(len(df_all)):
    wt_seq = df_all.loc[i, 'WildType.Sequence']
    mut_seq = df_all.loc[i, 'Mutant.Sequence']

    if wt_seq == mut_seq:
        m_type = "None"
    elif len(wt_seq) == len(mut_seq):
        m_type = "Substitution"
    elif len(wt_seq) > len(mut_seq):
        m_type = "Deletion"
    else:
        m_type = "Insertion"

    if wt_seq[:1000] != mut_seq[:1000]:
        m_region = "Promoter"
    else:
        m_region = "Coding"

    mutation_types.append(m_type)
    mutation_regions.append(m_region)
```

```python
df_all['Mutation.Type'] = mutation_types
df_all['Mutation.Region'] = mutation_regions
```

```python
df_all[['Gene', 'Mutation.Type', 'Mutation.Region']].head()
```

```python
df_all['Score'] = df_all['mRNA_change'].abs() + df_all['Protein_change'].abs() + df_all['Viability_change'].abs()
```

```python
df_top5 = df_all.sort_values(by='Score', ascending=False).head(5)
```
